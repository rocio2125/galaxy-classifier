```python
import tensorflow as tf

model = tf.keras.models.load_model("modelo_galaxias.keras")

# Re-exportar sin optimizador ni estado de entrenamiento
model.save("modelo_galaxias_inference.keras", include_optimizer=False)
```

```python
model = tf.keras.models.load_model("modelo_galaxias_inference.keras")

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.allow_custom_ops = True
converter.optimizations = [tf.lite.Optimize.DEFAULT]

tflite_model = converter.convert()

with open("modelo_galaxias.tflite", "wb") as f:
    f.write(tflite_model)
```
